#  Entendedo o perfil de ataque nos USA 

1. Criar dataframe chamado shark_attack_usa que trara apenas os dados da tabela com o filtro na coluna Country igual a USA

```python
shark_attack_usa=shark_attack.loc[shark_attack['Country']=='USA']
shark_attack_usa
```

2. Contar a quantidade de ataques nos USA pelo tipo de atividade, coluna Activity

```python
shark_attack_usa['Activity'].value_counts()
```

3. Conferindo o tamanho da tabela shark_attack_usa

# E nos outros países qual é o perfil de ataque de tubarões?

1. Criar um novo dataframe chamado shark_attack_without_usa com os ataques de todos os paises exceto nos EUA 

```python
shark_attack_without_usa=shark_attack.loc[shark_attack['Country']!='USA']
shark_attack_without_usa
```

2. Contar a quantidade de ataques nos USA pelo tipo de atividade, coluna Activity

```python
shark_attack_without_usa['Activity'].value_counts()
```

3. Conferindo o tamanho da tabela shark_attack_without_usa

```python
shark_attack_without_usa.shape
```

```python
4. Contar a quantidade de ataques de todos os outros paises exceto dos EUA, atraves do dataframe shark_attack_without_usa pelo tipo de ataque, coluna Type
```

```python
shark_attack_without_usa['Type'].value_counts()
```

# Entendendo os dados de forma consolidada

```python
1. Criação de grafico com as 5 maiores quantidades do tipo de atividade registradas no ataque de tubaroes nos USA 
```

```python
shark_attack_usa['Activity'].value_counts().head().plot.bar()
```

```python
2. Criação de grafico com as 5 maiores quantidades do tipo de atividade registradas no ataque de tubaroes nos outros paises
```

```python
shark_attack_without_usa['Activity'].value_counts().head().plot.bar()
```

```python
3. Criação de um gráfico consolidando os resultados das duas regiões
```

```python
pd.merge(shark_attack_usa['Activity'].value_counts().head(),shark_attack_without_usa['Activity'].value_counts(),left_index=True,right_index=True).plot.bar()
plt.legend(["USA", "Other Countries"])
```
